```python
import h5py
import glob
import fitsio
import numpy as np
import os
```

```python
fname = "metadetect_cutsv6_blinded.h5"
```

```python
n_load = 100_000
i = 0

with h5py.File(fname, "r") as fp:
    ra = fp["mdet/noshear/ra"][i * n_load:(i+1) * n_load]
    dec = fp["mdet/noshear/dec"][i * n_load:(i+1) * n_load]

    msk, = np.where((ra == 0) & (dec == 0))
    data = {}
    for col in fp["mdet/noshear"].keys():
        data[col] = fp["mdet/noshear"][col][msk + i * n_load]
```
