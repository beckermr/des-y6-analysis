```python
import h5py
import fitsio
import numpy as np
```

```python
mastercat = "metadetect_2024-11-07.hdf5"
final_hdf5cat = "../2023_12_21_v6cuts_hdf5cat/metadetect_cutsv6_all.h5"
mdet_files = "../2023_12_21_v6cuts_hdf5cat/mdet_flist.txt"
```

```python
mdet_fnames = []
with open(mdet_files, "r") as fp:
    for line in fp.readlines():
        mdet_fnames.append(line.strip())
```

# Check Final Catalog

```python
fp = h5py.File(mastercat, "r")
```

```python
fp["noshear"]["tomo_bin_0"].keys()
```

```python
fp.close()
```

```python
rng = np.random.RandomState(seed=42)

for _ in range(10):
    findex = rng.choice(len(mdet_fnames))
    print("file index:", findex)

    orig_data = fitsio.read(mdet_fnames[findex])
    print('    read fits data')

    with h5py.File(mastercat, "r") as fp:
        for key in fp.keys():
            for tkey in fp[key].keys():
                msk = fp[key][tkey]['tilename'][:] == orig_data['tilename'][0].encode('utf-8')
                ra = fp[key][tkey]['ra'][msk]
                dec = fp[key][tkey]['dec'][msk]
                g1 = fp[key][tkey]['gauss_g_1'][msk]
                g2 = fp[key][tkey]['gauss_g_2'][msk]

                print('    read hdf5 data')

                rind = rng.choice(ra.shape[0])

                rmsk = (
                    (orig_data['ra'] == ra[rind])
                    & (orig_data['dec'] == dec[rind])
                    & (orig_data['mdet_step'] == key)
                )
                assert np.sum(rmsk) == 1
                assert orig_data['gauss_g_1'][rmsk] == g1[rind]
                assert orig_data['gauss_g_2'][rmsk] == g2[rind]

                print('    did comp')
```
