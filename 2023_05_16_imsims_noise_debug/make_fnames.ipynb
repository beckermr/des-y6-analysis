```python
import fitsio
```

```python
d = fitsio.read("fnames.fits")

tnames = set()
for i in range(len(d)):
    _tname = d["FILENAME"][i].split("_")[0]
    tnames.add(_tname)
```

```python
tnames = sorted(tnames)
```

```python
import numpy as np
```

```python
rng = np.random.RandomState(seed=10)
_rtnames = rng.choice(tnames, size=100)
rng = np.random.RandomState(seed=10)
__rtnames = rng.choice(tnames, size=519)
rtnames = set(_rtnames.tolist() + __rtnames.tolist())
```

```python
print(len(rtnames))
```

```python
with open("tiles.txt", "w") as fp:
    for tname in rtnames:
        fp.write("%s\n" % tname)
```

```python
import joblib
```

```python
import subprocess
```

```python
def _down(tname):
    subprocess.run("./run_downloading.sh %s" % tname, check=True, shell=True, capture_output=True)
```

```python
jobs = [
    joblib.delayed(_down)(tname)
    for tname in rtnames
]

with joblib.Parallel(n_jobs=10, backend="threading", verbose=100) as par:
    par(jobs)
```
