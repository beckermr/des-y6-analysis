```python
import fitsio
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
d = fitsio.read("mdet_data/DES0240-2458_metadetect-v6_mdetcat_part0000.fits.fz", columns=[
    "mdet_step", "flags", "mdet_T_ratio", "mdet_s2n",
])
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1)

msk = d["flags"] == 0
# axs.hexbin(np.log10(d["mdet_s2n"][msk]), d["mdet_T_ratio"][msk], extent=(1, 5, -2, 2))
axs.plot(np.log10(d["mdet_s2n"][msk]), d["mdet_T_ratio"][msk], '.')
axs.set_ylim(0, 2.25)
# axs.axhline(1, color='k', linestyle="--")
axs.set_xlabel("log10[S/N]")
axs.set_ylabel("$T/T_{PSF}$")
```
